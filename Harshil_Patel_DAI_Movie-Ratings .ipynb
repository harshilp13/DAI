```python
#Q1 : List of top 10 movies in terms of budget(descending order)
df = pd.DataFrame(mydata, columns=['Film','Genre','Rotten Tomatoes Ratings %','Audience Ratings %','Budget (million $)','Year of release'])
df.sort_values(by=['Budget (million $)'], ascending=False).head(10)
#head(n) will give us the top n values from data 
#data frames and series are fundamental part of pandas
```

```python
#Q2 : List of top 10 movies in terms of budget(ascending order)
df.sort_values(by=['Budget (million $)'], ascending=True).head(10)
```

```python
#Q3 : Which genre is best in terms of price budget 
mydata = pd.read_csv(r"C:\Users\harshil\Desktop\DAI\Movie-Ratings.csv")
mydata.groupby('Genre').mean().drop('Year of release',axis=1).sort_values('Budget (million $)',ascending=True)
mydata.groupby('Genre').max()
#axis(0-index,1-columns) where default is 0
```

```python
#Q4 : Distribution graph of audience rating
sns.distplot(a=mydata['Audience Ratings %'])
#seaborn has already been imported in first command as sns
```

```python
#Q5 : Distribution graph of critics rating(rotten tomatoes)
sns.distplot(a=mydata['Rotten Tomatoes Ratings %'])
#On analysis, of Q4 and 5, we can see the difference in ratings for genres by audience and critics
```

```python
#Q6 : Distribution of budget
sns.boxplot(x=mydata['Genre'],y=mydata['Budget (million $)'],hue=mydata['Year of release'])
#By analysis of bar graph, we can see that the distibution of budget for movies is majorly between $ 0 and 50 million 
#bins and kde can be set for better understanding by setting bins=100 or more and kde=False
```

```python
#Q7 : Distribution of critics rating
sns.boxplot(x=mydata['Genre'],y=mydata['Rotten Tomatoes Ratings %'],hue=mydata['Year of release'])
```

```python
#Q8 : Distribution of audience rating
sns.boxplot(x=mydata['Genre'],y=mydata['Audience Ratings %'],hue=mydata['Year of release'])
#On analysis, of Q7 and 8, we can see the difference in distribution of ratings by critics and audience
```
